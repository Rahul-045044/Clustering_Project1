## Dataset Description:

### Valence:

Valence represents a measure of the musical positiveness conveyed by a track, ranging from 0 to 1. A value closer to 1 indicates a more positive emotional tone, while a value closer to 0 represents a less positive tone.

### Year:

Year denotes the release year of the track, providing temporal information about the music's historical context and evolution over time.

### Acousticness:

Acousticness quantifies the acoustic characteristics of the track on a scale from 0 to 1. A value of 0 suggests a predominantly electronic or synthesized sound, while a value of 1 indicates a purely acoustic recording.

###Danceability:

Danceability measures how suitable a track is for dancing, ranging from 0 to 1. Higher values imply tracks with rhythms and beats conducive to dancing.

###Duration_ms:

Duration_ms specifies the duration of the track in milliseconds, providing insight into the length of the musical composition.

###Energy:

Energy reflects the intensity and vigor of the track's musical elements, ranging from 0 to 1. Higher values indicate more energetic and lively compositions.
###Instrumentalness:

Instrumentalness quantifies the likelihood that the track contains no vocals, ranging from 0 to 1. A higher value suggests a higher probability of instrumental-only compositions.

###Liveness:

Liveness measures the presence of a live audience in the recording, with values ranging from 0 to 1. A higher value indicates a higher probability of a live performance.

###Loudness:

Loudness denotes the overall loudness of the track in decibels (dB), providing information about its volume and intensity.

###Popularity:

Popularity indicates the popularity of the track, reflecting its reception and engagement among listeners.

###Speechiness:

Speechiness measures the presence of spoken words in the track, ranging from 0 to 1. Higher values indicate tracks with more spoken-word elements, such as podcasts or spoken-word performances.

###Tempo:

Tempo represents the speed or pace of the track in beats per minute (BPM), providing information about its rhythmic structure and tempo.
